```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
with open("allData.pkl","rb") as f:
    data=pickle.load(f)
```

```python
data.keys()
```

```python
#View the data
data['TomQuality']
```

```python
#storing files under TomQuality to tom
tom = data['TomQuality']
```

```python
#Viewing the shape of data
tom.shape
```

```python
#Viewing data info
tom.info()
```

```python
#drop unsuitable data columns
tom=tom.drop(['\tFlavour', ' \tTSS', '\tAcid', '\t%Juice', '\tBite','\tWeight\tDMC_fruit'], axis=1)
```

```python
#Viewing the shape of data after droping 
tom.shape
```

```python
#Describing the data
tom.describe()
```

```python
#Checking for number of unique data entries in each column
tom.nunique()
```

```python
#Checking for unique data entries in each column
for col in tom.columns:
    print('{} : {}'.format(col,tom[col].unique()))
```

```python
#Checking the data type of each column
tom.dtypes
```

```python
#Displaying all the column headings
tom.columns
```

```python
#Renaming
tom.rename(columns={'%time': 'Time'}, inplace=True)
```

```python
#Viewing the data
tom.head()
```

```python
#changing time to its standard format 
tom['Time'] = pd.to_datetime(tom['Time'], unit='D', origin='1899-12-30')
```

```python
#View the data
tom.head()
```

```python
#Checking for Null values in the data
tom.isnull()
```

```python
#Checking for the sum of the null values in the data
tom.isnull().sum()
```

```python
#Visualizing the null values
sns.heatmap(tom.isnull(),cbar=False,cmap='inferno')
```

```python
#Replacing the null values with mean of their respective columns
num_col = ['Flavour ', 'TSS', 'Acid', '%Juice', 'Bite', 'Weight', 'DMC_fruit']
for col in num_col:
    tom[col]=pd.to_numeric(tom[col], errors= 'coerce')
    tom[col].fillna(tom[col].mean(), inplace=True)
```

```python
#Visualizing the null values again after replacing 
sns.heatmap(tom.isnull(),cbar=False,cmap='inferno')
```

```python
#Again checking for the null values sum in the data 
tom.isnull().sum()
```

```python
#Getting all the column headings in a list
list(tom.columns.values)
```

```python
#Storing all the listed column headings excluding 'Time' in  a variable 
list_tom=list(tom.columns.values)
list_tom.remove('Time')
```

```python
#Converting all the int data types to float 
for i in list_tom:
    tom[i] = tom[i].astype(float)
```

```python
#Printing data types of each column
tom.dtypes
```

```python
#Finding Correlation
correlation=tom.corr()
correlation
```

# Plotting Line Plot to analyze the data features with respect to time

```python
for col in ['Flavour ', 'TSS', 'Acid', '%Juice', 'Bite', 'Weight', 'DMC_fruit']:
    plt.figure(figsize=(10,6))
    sns.lineplot(data=tom, x='Time', y=col)
```
