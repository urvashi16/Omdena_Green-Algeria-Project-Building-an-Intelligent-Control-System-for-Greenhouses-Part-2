```python
#Renaming
data.rename(columns={"%time": "Time"}, inplace=True)
```

```python
#View the data
data.head()
```

```python
#changing time to its standard format 
data['Time'] = pd.to_datetime(data['Time'], unit='D', origin='1899-12-30')
```

```python
#View the data
data.head()
```

```python
#Checking for Null values in the data
data.isnull()
```

```python
#Checking for the sum of the null values in the data
data.isna().sum()
```

```python
#Replacing the null values with mean of their respective columns
num_col = ['AbsHumOut','Iglob','PARout','Pyrgeo','RadSum','Rain','Rhout','Tout','Winddir','Windsp']
for col in num_col:
    data[col]=pd.to_numeric(data[col])
    data[col].fillna(data[col].mean(), inplace=True)
```

```python
#Again checking for the null values sum in the data 
data.isnull().sum()
```

```python
#Getting all the column headings in a list
list(data.columns.values)
```

```python
#Storing all the listed column headings excluding 'Time' in  a variable 
list_data=list(data.columns.values)
list_data.remove('Time')
```

```python
#Converting all the int data types to float 
for i in list_data:
    data[i] = data[i].astype(float)
```
